```python

import cv2
import numpy as np
from matplotlib import pyplot as plt
image = cv2.imread('space.png')

if image is None:
    print("Error: Unable to load the image.")
else:
    gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
plt.imshow(image)
plt.axis('off')
plt.show()


pixel_values, counts = np.unique(gray_image, return_counts=True)
total_pixels = np.sum(counts)
probabilities = counts / total_pixels

symbol_probabilities = {}
for i, pixel_value in enumerate(pixel_values):
    symbol_probabilities[f's{i+1}'] = probabilities[i]
assert np.isclose(sum(symbol_probabilities.values()), 1.0), "Sum of probabilities is not 1"
print("Symbol Probabilities:")
for symbol, probability in symbol_probabilities.items():
    print(f"{symbol}: {probability}")
```
